```python
import time
import numpy as np
import matplotlib.pyplot as plt

from bayegent import Bayegent
from environment import GridMazeEnvironment
```

```python
def run_one_seed(seed, parameters, n_runs=100):
    environment = GridMazeEnvironment(seed)
    agent = Bayegent(environment, seed, parameters=parameters)

    position_histories = agent.learn_bayesian(n_runs)

    return position_histories
```

## Sweep over static curiosity values

For each curiosity value between 0.1 and 0.9, spaced by 0.02, check what the average path length is for all of the first 100 seeds


```python
curiosity_values = np.linspace(0.5,0.9,15)
mean_path_lengths = {c: 0 for c in curiosity_values}

# Iterate over curiosity values
for curiosity in curiosity_values:
    print(f'Curiosity: {curiosity}')
    path_lengths = []
    # Iterate over 100 seeds
    for seed in range(50):
        print(f'{seed}/50')
        position_history = run_one_seed(seed, 
                                        parameters={
                                            'curiosity': curiosity,
                                            'step_reward': -0.1,
                                            'goal_reward': 1,
                                            'learning_rate': 0.5,
                                            'discount_factor': 0.8,
                                        },
                                        n_runs=100)

        path_lengths.append(len(position_history))
    
    mean_path_length = np.mean(path_lengths)
    mean_path_lengths[curiosity] = mean_path_length
```

```python
c = mean_path_lengths.keys()
mpl = mean_path_lengths.values()

plt.plot(c, mpl)
```

```python
curiosity_values = np.linspace(0.3,0.9,7)

all_last_path_lengths = {c: 0 for c in curiosity_values}
all_running_times = {c: 0 for c in curiosity_values}
all_min_path_lengths = {c: 0 for c in curiosity_values}

mean_last_path_lengths = {c: 0 for c in curiosity_values}
mean_running_times = {c: 0 for c in curiosity_values}
mean_min_path_lengths = {c: 0 for c in curiosity_values}

# Iterate over curiosity values
for curiosity in curiosity_values:
    print(f'Curiosity: {curiosity}')
    last_path_lengths = []
    min_path_lengths = []
    running_times = []

    # Iterate over 100 seeds
    for seed in range(20):
        print(f'{seed}/20')
        start = time.time()
        position_histories = run_one_seed(seed, 
                                        parameters={
                                            'curiosity': [curiosity for _ in range(100)],
                                            'step_reward': -0.1,
                                            'goal_reward': 1,
                                            'learning_rate': 0.5,
                                            'discount_factor': 0.8,
                                        },
                                        n_runs=100)
        end = time.time()

        path_lengths = [len(pl) for pl in position_histories]

        running_times.append(end-start)
        last_path_lengths.append(path_lengths[-1])
        min_path_lengths.append(min(path_lengths))
    
    all_last_path_lengths[curiosity] = last_path_lengths
    all_min_path_lengths[curiosity] = min_path_lengths
    all_running_times[curiosity] = running_times

    mean_last_path_lengths[curiosity] = np.mean(last_path_lengths)
    mean_min_path_lengths[curiosity] = np.mean(min_path_lengths)
    mean_running_times[curiosity] = np.mean(running_times)
```

## Running Time

To show that we can't just choose a low curiosity, we need to show the learning time for lower curiosities are much much higher

```python
curiosity_values = np.linspace(0.3,0.9,7)
running_times = {c: 0 for c in curiosity_values}

# Iterate over curiosity values
for curiosity in curiosity_values:
    print(f'Curiosity: {curiosity}')
    times = []
    # Iterate over 100 seeds
    for seed in range(50):
        print(f'{seed}/50')
        start = time.time()
        position_history = run_one_seed(seed, 
                                        parameters={
                                            'curiosity': curiosity,
                                            'step_reward': -0.1,
                                            'goal_reward': 1,
                                            'learning_rate': 0.5,
                                            'discount_factor': 0.8,
                                        },
                                        n_runs=100)

        end = time.time()

        times.append(end - start)

    running_times[curiosity] = times
```

```python
c = sorted(running_times.keys())
rt = [np.mean(running_times[v]) for v in c]

plt.plot(c, rt)
plt.xlabel('Curiosity')
plt.ylabel('Mean Learning Time (sec)')
plt.title('Average Learning Time for 100 runs across 100 seeds')

plt.axhline(y=8.95, color='green', label='Linear weaning 0.9 to 0.2', linestyle=':')
plt.legend()
```

## Weaning curiosity

Try the first 100 with a weaning curiosity, weaning from 0.9 to 0.3 over 100 runs

```python
curiosity_values = np.linspace(0.3,0.9,7)
all_min_path_lengths = {c: None for c in curiosity_values}
mean_min_path_lengths = {c: None for c in curiosity_values}

# Iterate over curiosity values
for curiosity in curiosity_values:
    print(f'Curiosity: {curiosity}')
    path_lengths = []
    # Iterate over 10 seeds
    for seed in range(10):
        print(f'{seed}/10')
        position_histories = run_one_seed(seed, 
                                        parameters={
                                            'curiosity': [curiosity for _ in range(100)],
                                            'step_reward': -0.1,
                                            'goal_reward': 1,
                                            'learning_rate': 0.5,
                                            'discount_factor': 0.8,
                                        },
                                        n_runs=100)

        pls = [len(ph) for ph in position_histories]
        path_lengths.append(pls)
    
    min_path_lengths = [np.min(pls) for pls in path_lengths]
    all_min_path_lengths[curiosity] = path_lengths
    mean_min_path_lengths[curiosity] = min_path_lengths
```
